```python
import pandas as pd
import numpy as np

df_a=pd.read_csv('data/stockinfo.itemname.csv',encoding='cp949')
df_b=pd.read_csv('data/stock.adj_close.csv',encoding='cp949')
```

```python
df_a.head()
```

```python
df_b.head()
```

```python
# 여러개의 DataFrame 합치기
#concat() : 여러 개의 데이터프레임 합치는 함수
df_a=df_a.iloc[:14]
df_a
```

```python
df_b=df_b.iloc[:14]
df_b
```

```python
# axis= 키워드 인자로 병합하는 방향 지정
# 0: 행 방향 (아래방향)
# 1: 열 방향 (->)
# pd.concat([df_a,df_b]) : 행 방향으로 병합
# pd.concat([df_a,df_b],axis=1) : 열 방향으로 병합
pd.concat([df_a,df_b])
```

```python
pd.concat([df_a,df_b], axis=1)
```
