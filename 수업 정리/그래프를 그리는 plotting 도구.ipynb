```python
# seaborn 으로 데이터 분포 살펴보기
# 데이터의 통계적인 부분을 살펴볼 때, matplotlib에 비해 쉽고 간편함.

import seaborn as sns
sns.set()
plt.rc('font',family='Malgun Gothic')
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data=tips,x='day')
plt.title('요일별 팁을 준 횟수')
```

```python
# 사분위 도표
sns.boxplot(data=tips,x='day',y='tip')
plt.title('요일별 팁의 Box Plot')

# |0%,ㅡ25%,50%,75%,ㅡ100%| 
```

```python
# 두 변수 관계를 점찍어 그리기
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.title('식사 금액과 팁의 관계')
```

```python
# 두변수 관계를 점과 분포로 보기
sns.jointplot(data=tips,x='total_bill',y='tip')
plt.suptitle('식사 금액과 팁의 관계',y=1.02)
```

```python
# hue 인자로 그래프 분리하기
sns.boxplot(data=tips,x='day',y='tip',hue='sex')
plt.title('요일 별, 성별 팁의 Box Plot')
```

```python
# 빈도를 바이올린 모양으로 viloinplot
sns.violinplot(data=tips,x='day',y='tip',hue='sex')
plt.title('요일별, 성별 팁의 violinplot')
plt.show()
```

```python
sns.displot(tips.tip, kde=True,rug=True)
plt.title('팁의 분포')
```
